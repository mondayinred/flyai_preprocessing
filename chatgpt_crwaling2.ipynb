```python
import os
import time
import random
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile
from selenium.webdriver.common.by import By

# DISPLAY 환경 변수 설정 (Windows의 IP 주소 사용)
os.environ["DISPLAY"] = "172.21.240.1:0"  # 여기서 '192.168.x.x'는 Windows IP 주소로 바꿔야 합니다.

# Firefox 옵션 설정
options = Options()
options.headless = False
options.set_preference("marionette", False)
options.set_preference("dom.webdriver.enabled", False)
options.set_preference("useAutomationExtension", False)
options.set_preference("general.useragent.override", "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")
options.set_preference("intl.charset.default", "UTF-8")

# Firefox 실행
driver = webdriver.Firefox(options=options)
driver.set_window_size(1920, 1080)
time.sleep(random.uniform(2, 4))
driver.maximize_window()
# driver.execute_script("""
#     var req = new XMLHttpRequest();
#     req.open('GET', window.location, true);
#     req.setRequestHeader('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36');
#     req.send(null);
# """)
driver.execute_script("""
    var meta = document.createElement('meta');
    meta.setAttribute('charset', 'UTF-8');
    document.getElementsByTagName('head')[0].appendChild(meta);
""")
# driver.execute_script("""
#     Object.defineProperty(navigator, 'webdriver', { get: () => undefined });
#     Object.defineProperty(navigator, 'languages', { get: () => ['en-US', 'en'] });
#     Object.defineProperty(navigator, 'plugins', { get: () => [{ name: 'Chrome PDF Plugin', filename: 'pdf' }] });
# """)
driver.get('https://chat.openai.com')

time.sleep(random.uniform(3, 4))

try:
    # 텍스트 입력
    prompt = driver.find_element(By.XPATH, '//*[@id="prompt-textarea"]/p')
    print(prompt.is_enabled())
    prompt.send_keys('웹 크롤링이 뭐야?')
    time.sleep(2)
    
    # 버튼 클릭
    button = driver.find_element(By.XPATH, '//*[@id="composer-background"]/div[2]/div[2]/div/button')
    # button = driver.find_element(By.XPATH, '//*[@id="search-btn"]')
    print(button.is_enabled())
    button.click()
    time.sleep(random.uniform(2, 3))
except Exception as e:
    print('error')
    
time.sleep(10)
```
