```python
import json
import numpy as np
import pandas as pd

file_path = "/home/kyumin/work/preprocessing/datasets/BF-PSR-Framework/JsonData/PJZ.txt"

df = pd.DataFrame()
with open(file_path) as json_file:
    conversations = json.load(json_file)
    for p in conversations['conversation']:
        messages = p['messages']
        df_message = pd.DataFrame(messages)
        m_id = 0
        df_temp = pd.DataFrame()
        # print(df_message.shape[0])
        for j in range(df_message.shape[0]): 
            df_temp.loc[j, 'source'] = p['source']
            df_temp.loc[j, 'label'] = p['label']
            df_temp.loc[j, 'message_idx'] = p['id']
        df_temp = pd.concat([df_temp, df_message], axis=1)
        # print(df_temp.shape)
        df = pd.concat([df, df_temp], axis=0)
        # print(df.shape)
        m_id += 1
        

df
```

```python
df['time'] = df['time'].apply(str)
df['time'] = df['time'].apply(lambda x: f'"{x}"') # 이거 안하면 엑셀에서 이상하게 읽어버림림
df['message_idx'] = df['message_idx'].apply(int)
df
```

```python
df.to_csv("PJZ2.csv", index=False, encoding='utf-8')
```
